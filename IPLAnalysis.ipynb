```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/Rahul Negi/Downloads/IPL Dataset.csv")
df
```

```python
plt.figure(figsize=(10,8))
match_played_per_year= df.groupby(['match_id','season']).count().index.droplevel(level=0).value_counts().sort_index()
sns.barplot(y=match_played_per_year.index,x=match_played_per_year,orient='h',width=0.5)
plt.xlabel('match played')
plt.ylabel('season')
plt.show()
```

```python
plt.figure(figsize=(10,8))
match_played_at_venue= df.groupby(['match_id','venue']).count().index.droplevel(level=0).value_counts().sort_index()
sns.barplot(y=match_played_at_venue.index,x=match_played_at_venue,orient='h')
plt.xlabel('match played')
plt.ylabel('venue')
plt.show()
```

```python
match_played_at_venue= df.groupby(['match_id','venue']).count().index.droplevel(level=0).value_counts()
match_played_at_venue
```

```python
plt.figure(figsize=(10,8))
match_played_by_team= df.groupby(['match_id','bowling_team']).count().index.droplevel(level=0).value_counts().sort_values(ascending=False)
sns.barplot(y=match_played_by_team.index,x=match_played_by_team,orient='h',width=0.5)
plt.xlabel('match played')
plt.ylabel('team')
plt.show()
```

```python
most_runs_by_teams= df.groupby(['batting_team'])['run'].sum().sort_values(ascending=False)
most_runs_by_teams
```

```python
most_runs_by_player= df.groupby(['striker'])['runs_off_bat'].sum().sort_values(ascending=False)[:10]
most_runs_by_player
```

```python
plt.figure(figsize=(10,8))
most_runs_by_player=df.groupby(['striker'])['runs_off_bat'].sum().sort_values(ascending=False)[:10]
sns.barplot(y=most_runs_by_player.index,x=most_runs_by_player,orient='h',width=0.5)
plt.xlabel('match played')
plt.ylabel('team')
plt.show()
```

```python
best_average_in_powerplay=df[df['over']<=6].groupby(['match_id','batting_team']).sum()['run'].groupby('batting_team').mean().sort_values(ascending=False)[2:]
best_average_in_powerplay
```

```python
runs_by_batsman= df.groupby(['match_id','striker'])['runs_off_bat'].sum()
fifty_by_batsman=runs_by_batsman[runs_by_batsman>=50]
most_fifty_by_batsman=fifty_by_batsman.groupby('striker').count().sort_values(ascending=False)[:10]
most_fifty_by_batsman
```

```python
runs_by_batsman= df.groupby(['match_id','striker'])['runs_off_bat'].sum()
hundred_by_batsman=runs_by_batsman[runs_by_batsman>=100]
most_hundred_by_batsman=hundred_by_batsman.groupby('striker').count().sort_values(ascending=False)[:10]
most_hundred_by_batsman
```

```python
Ist='caught,bowled,lbw,stumped,caught and bowled,hit wicket'
most_wicket_by_bowler=df[df['wicket_type'].apply(lambda x: True if x in Ist and x != ' ' else False)]['bowler'].value_counts()[:10]
most_wicket_by_bowler
```

```python
most_dot_balls_by_bowler=df[df['run']==0].groupby(['bowler']).count()['match_id'].sort_values(ascending=False)[:10]
most_dot_balls_by_bowler
```

```python
most_six_by_batsman_in_single_inning=df[df['run']==6].groupby(['start_date','striker']).count()['season'].sort_values(ascending=False)[:10]
most_six_by_batsman_in_single_inning
```

```python
most_four_in_single_inning=df[df['run']==4].groupby(['start_date','striker']).count()['season'].sort_values(ascending=False)[:10]
most_four_in_single_inning
```

```python
most_run_in_a_season=df.groupby(['striker','season'])['runs_off_bat'].sum().sort_values(ascending=False)[:10]
most_run_in_a_season
```

```python
highest_score_by_team=df.groupby(['start_date','batting_team']).sum()['run'].droplevel(level=0).sort_values(ascending=False)[:10]
highest_score_by_team
```

```python
highest_score_by_batsman=df.groupby(['start_date','striker']).sum()['runs_off_bat'].sort_values(ascending=False)[:10]
highest_score_by_batsman
```

```python
most_run_conceded_by_bowler=df.groupby(['start_date','bowler']).sum()['run'].sort_values(ascending=False)[:10]
most_run_conceded_by_bowler
```

```python
run_by_bowler=df.groupby(['start_date','bowler','over'])['run'].sum()
maiden_over_by_bowler=run_by_bowler[run_by_bowler.values == 0].droplevel(level=[0,2])
maiden_over_by_bowler.index.value_counts()[:10]
```

```python
ist='caught,bowled,lbw,stumped,caught and bowled,hit wicket'
most_wicket_taken_by_team=df[df['wicket_type'].apply(lambda x: True if x in ist and x != ' ' else False)]['bowling_team'].value_counts()
most_wicket_taken_by_team
```

```python
most_extras_by_team=df.groupby(['bowling_team'])['extras'].agg('sum').sort_values(ascending=False)
most_extras_by_team
```
